# TASK 3:Exploratory Data Analysis on dataset SampleSuperstore

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "C:/Users/Harishma/Downloads/Internship/Task 3/SampleSuperstore.csv"
data = pd.read_csv(file_path)
data
```

```python
print(data.head())
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
# No missing values, data types look appropriate
```

```python
print(data.describe())
```

```python
# Sales and Profit Analysis
# Sales and Profit by Region
sales_by_region = data.groupby('Region')['Sales'].sum().reset_index()
profit_by_region = data.groupby('Region')['Profit'].sum().reset_index()
```

```python
# Sales and Profit by Category
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
```

```python
# Sales and Profit by Sub-Category
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
```

```python
# Sales and Profit by Segment
sales_by_segment = data.groupby('Segment')['Sales'].sum().reset_index()
profit_by_segment = data.groupby('Segment')['Profit'].sum().reset_index()
```

# VISUALIZATION

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Sales', data=sales_by_region)
plt.title('Total Sales by Region')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Profit', data=profit_by_region)
plt.title('Total Profit by Region')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Sales', data=sales_by_category)
plt.title('Total Sales by Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Profit', data=profit_by_category)
plt.title('Total Profit by Category')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Sub-Category', y='Sales', data=sales_by_subcategory)
plt.xticks(rotation=90)
plt.title('Total Sales by Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Sub-Category', y='Profit', data=profit_by_subcategory)
plt.xticks(rotation=90)
plt.title('Total Profit by Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Segment', y='Sales', data=sales_by_segment)
plt.title('Total Sales by Segment')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Segment', y='Profit', data=profit_by_segment)
plt.title('Total Profit by Segment')
plt.show()
```

```python
# Discount Analysis
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Discount', y='Sales', data=data)
plt.title('Sales vs Discount')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Discount', y='Profit', data=data)
plt.title('Profit vs Discount')
plt.show()
```

```python
# Category-wise Sales
category_region = data.groupby(['Category', 'Region'])['Sales'].sum().unstack()
plt.figure(figsize=(12, 7))
category_region.plot(kind='bar', stacked=True)
plt.title('Category-wise Sales by Region')
plt.show()
```

```python
# Heatmap for Sales and Profit by Region and Category
pivot_sales = data.pivot_table(values='Sales', index='Region', columns='Category', aggfunc='sum')
pivot_profit = data.pivot_table(values='Profit', index='Region', columns='Category', aggfunc='sum')
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(pivot_sales, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Sales Heatmap by Region and Category')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(pivot_profit, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Profit Heatmap by Region and Category')
plt.show()
```
